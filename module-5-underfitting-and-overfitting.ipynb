**Solutions to Underfitting and Overfitting: A Comprehensive Overview**

Before we dive into the code, let's outline the **major strategies** we will explore in this module to address underfitting and overfitting. Understanding these solutions conceptually will provide a roadmap for our practical exercises.

**1. Solutions for Underfitting (Model is Too Simple):**

* **Increase Model Complexity:**
    * **For Decision Trees:** Increase `max_depth`, decrease `min_samples_split`, decrease `min_samples_leaf`, increase `max_leaf_nodes`. Allow the tree to grow more and capture more intricate patterns.
    * **For Linear Models:** Consider using more complex models, like Polynomial Regression (though we are not using it in this revised module for simplicity, it's a conceptual option), or more flexible models like Neural Networks (beyond the scope of this module, but conceptually relevant).
* **Feature Engineering:** Create new, more informative features from existing ones. This can provide the model with richer input data to learn from. (We won't focus heavily on feature engineering in *this* module, but it's a crucial concept to remember).
* **Reduce Regularization:** If you have applied regularization techniques (which are primarily for overfitting), reducing or removing them can sometimes help an underfitting model become more complex.

**2. Solutions for Overfitting (Model is Too Complex):**

* **Decrease Model Complexity:**
    * **For Decision Trees:** Decrease `max_depth`, increase `min_samples_split`, increase `min_samples_leaf`, decrease `max_leaf_nodes`.  Prune the tree to make it simpler and less prone to memorizing noise.
    * **For Linear Models:**  Use simpler linear models if possible.
* **Regularization Techniques:** Apply techniques that constrain model complexity and penalize overly complex models.
    * **For Decision Trees:**  Use parameters like `max_depth`, `min_samples_split`, `min_samples_leaf`, `max_leaf_nodes`, and `ccp_alpha` (Cost Complexity Pruning - more advanced).
    * **For Linear Models (like Logistic Regression - though not explicitly shown in *this* module's code):** L1 and L2 regularization (e.g., using parameters like `penalty='l1'` or `penalty='l2'` and adjusting the regularization strength parameter `C` in Logistic Regression).
* **Increase Training Data:**  Providing more training data can often help an overfitting model generalize better. With more examples, the model is less likely to memorize noise specific to a small training set.
* **Feature Selection / Feature Reduction:** Reduce the number of input features. If you have many irrelevant or noisy features, they can contribute to overfitting. Feature selection techniques (choosing the most relevant features) or feature reduction techniques (like Principal Component Analysis - PCA, beyond the scope of this module) can be helpful.
* **Cross-Validation:**  Use techniques like k-fold cross-validation to get a more robust estimate of model performance and to tune hyperparameters effectively. Cross-validation helps you avoid overfitting to a specific validation set. (We use a simple train-validation split in *this* module for simplicity, but cross-validation is crucial in practice).
* **Ensemble Methods:**  Use ensemble methods like Random Forests or Gradient Boosting (covered in Module 6). Ensemble methods combine multiple models, often Decision Trees, in a way that reduces overfitting and improves generalization.

### Comparing `max_depth`, `max_leaf_nodes`, and `min_samples_leaf` in Decision Trees

These three parameters are crucial for controlling the complexity and preventing overfitting in decision trees. Here's a breakdown:

| Parameter | Purpose | Effect | Typical Use |
|---|---|---|---|
| `max_depth` | Controls the maximum depth (number of levels) of the tree. | **Lower:** Simpler trees, potential underfitting. **Higher:** More complex trees, potential overfitting. | Limiting tree depth to prevent overfitting and improve generalization. |
| `max_leaf_nodes` | Limits the total number of leaf nodes (terminal nodes for predictions). | **Lower:** Restricts tree growth, simpler trees. **Higher:** Allows larger, more complex trees. | Controlling tree size to prevent overfitting, especially when `max_depth` is too restrictive. |
| `min_samples_leaf` | Sets the minimum number of samples required at a leaf node. | **Higher:** Prevents leaf nodes with very few samples, reducing overfitting. **Lower:** Allows more specific leaf nodes, potentially increasing complexity. | Smoothing decision boundaries and reducing overfitting by preventing the tree from fitting noise. |

**In essence, all three parameters are used for regularization:**

*   **`max_depth`** directly limits the depth or complexity of the tree.
*   **`max_leaf_nodes`** limits the number of leaf nodes or how much the tree can branch out.
*   **`min_samples_leaf`** ensures that leaf nodes represent a sufficient number of samples, preventing the creation of very specific and potentially overfit rules.

By adjusting these parameters, you can find the right balance between underfitting (model is too simple) and overfitting (model is too complex), leading to a well-generalizing decision tree model.

**Step 1: Prepare Data for Regression Task (House Price Prediction)**
---

Now that we have loaded the Iowa house price dataset, we need to prepare it specifically for our **regression task**, which is predicting house prices. This involves:

1.  **Selecting Target Variable:** Identifying the column we want to predict (SalePrice).
2.  **Selecting Features:** Choosing the columns we will use to make predictions. For simplicity in this module, we will focus on **numeric features** only.
3.  **Splitting Data:** Dividing the data into training and validation sets to properly evaluate our models.

```python
# Code Cell 5: Prepare Data - Regression Task (House Price Prediction)

# Print a header to indicate the start of Step 1: Prepare Data for Regression.
print("\n\n----- Step 1: Prepare Data for Regression -----")

# 1.1 Select the target variable for regression, which is 'SalePrice' from the iowa_data DataFrame.
# We want to predict the sale price of houses, so this is our target variable (y_regression).
y_regression = iowa_data['SalePrice']

# Print a message to confirm the creation of the target variable.
print("Regression Target variable 'y_regression' created!")

# Display the first 5 values of the target variable using y_regression.head().
print("\nFirst 5 values of the target variable 'y_regression':")
print(y_regression.head())
print(y_regression.info())
print(y_regression.describe())

# 1.2 Select numeric features for the regression task.
# We are choosing to use only numeric columns from the iowa_data DataFrame as features (X_regression).
numeric_features_regression = iowa_data.select_dtypes(include=np.number)

# Drop the 'SalePrice' column from the numeric features, as it is our target variable, not a feature.
# We want to use other numeric features to predict 'SalePrice'.
X_regression = numeric_features_regression.drop('SalePrice', axis=1)

# Print a message to confirm the creation of the feature matrix.
print("\nPredictive features 'X_regression' created!")

# Display the first 5 rows of the feature matrix using X_regression.head().
print("\nFirst 5 rows of regression features 'X_regression':")
print(X_regression.head())
print(X_regression.info())
print(X_regression.columns)

# 1.3 Split the data into training and validation sets for the regression task.
# We use train_test_split to split X_regression and y_regression into training and validation sets.
# test_size=0.2 means 20% of the data will be used for validation, and 80% for training.
# random_state=0 ensures that the split is reproducible.
X_train_reg, X_val_reg, y_train_reg, y_val_reg = train_test_split(X_regression, y_regression, test_size=0.2, random_state=0)

# Print a message to confirm that the regression data has been split into training and validation sets.
print("\nRegression data split into training and validation sets.")

print("Shape of X_train_reg:", X_train_reg.shape)
print("Shape of X_val_reg:", X_val_reg.shape)
print("Shape of y_train_reg:", y_train_reg.shape)
print("Shape of y_val_reg:", y_val_reg.shape)
```

**In essence, this cell prepares the data for our regression models by selecting the target and features, and then splitting the data into training and validation sets.  The training set will be used to train our models, and the validation set will be used to evaluate their performance and detect underfitting or overfitting.**

```python
# Code Cell 6: Regression Models - Training and Evaluation

# Print a header to indicate the start of Step 2: Regression Models - Training and Evaluation.
print("\n\n----- Step 2: Regression Models - Training and Evaluation -----")

# Define a dictionary 'regression_models' to store different regression models we want to evaluate.
# Each key is a descriptive name for the model, and each value is the model object itself.
regression_models = {
    "Decision Tree (max_depth=3) - Underfitting": DecisionTreeRegressor(max_depth=3, random_state=0), # Decision Tree with max_depth=3 (simple model, likely to underfit)
    "Decision Tree (No Limit) - Overfitting": DecisionTreeRegressor(random_state=0), # Decision Tree with no depth limit (complex model, likely to overfit)
    "Decision Tree (Regularized) - max_depth=10": DecisionTreeRegressor(max_depth=10, random_state=0), # Decision Tree with max_depth=10 (regularized model)
    "Linear Regression": LinearRegression() # Linear Regression model (simple model, might underfit)
}

# Initialize an empty dictionary 'regression_results' to store the Mean Absolute Error (MAE) for each model.
regression_results = {}

# Import SimpleImputer from sklearn.impute to handle missing values in the dataset.
from sklearn.impute import SimpleImputer

# Create an instance of SimpleImputer with the strategy='mean' to fill missing values with the mean of each column.
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training features (X_train_reg) and transform both training and validation features.
# fit_transform() learns the mean from the training data and applies imputation to the training data.
X_train_reg_imputed = imputer.fit_transform(X_train_reg)
# transform() applies the imputation learned from the training data to the validation data.
X_val_reg_imputed = imputer.transform(X_val_reg) # use transform on validation to avoid data leakage

# Loop through each model in the 'regression_models' dictionary.
for name, model in regression_models.items():
    # Train the current regression model using the imputed training features (X_train_reg_imputed) and training target (y_train_reg).
    model.fit(X_train_reg_imputed, y_train_reg) # Use imputed training data

    # Make predictions on the imputed validation features (X_val_reg_imputed) using the trained model.
    predictions = model.predict(X_val_reg_imputed) # Use imputed validation data

    # Calculate the Mean Absolute Error (MAE) by comparing the predicted values to the actual validation target values (y_val_reg).
    mae = mean_absolute_error(y_val_reg, predictions)

    # Store the calculated MAE in the 'regression_results' dictionary with the model name as the key.
    regression_results[name] = mae

    # Print the model name and its corresponding MAE value.
    print(f"Model: {name}, MAE: {mae:.2f}")

# Print a header to display the regression results in a table format.
print("\n--- Regression Model Performance (MAE) ---")

# Create a pandas DataFrame from the 'regression_results' dictionary to display the results in a table.
# The DataFrame will have model names as index and 'Validation MAE' as column name.
df_regression_results = pd.DataFrame.from_dict(regression_results, orient='index', columns=['Validation MAE'])

# Print the DataFrame containing the regression results table.
print(df_regression_results)
```

```python
# Code Cell 11: Visualize Classification Model Performance - ROC Curves (for models that can output probabilities) - Multiple Plots

# Print a header to indicate the start of Step 7: Visualize Classification Model Performance - ROC Curves.
print("\n\n----- Step 7: Visualize Classification Model Performance - ROC Curves -----")

# Define the probabilistic_models dictionary here
probabilistic_models = {
    "Logistic Regression (max_iter=1000)": LogisticRegression(max_iter=1000, random_state=0), # Example probabilistic model
    # Add other models that output probabilities here
}

# Fit the models before looping
for name, model in probabilistic_models.items():
    # Assuming you have X_train_clf and y_train_clf from previous cells
    model.fit(X_train_clf, y_train_clf)  # Fit the model here

# Get the number of probabilistic models to create subplots.
n_models = len(probabilistic_models)

# Create a figure and a set of subplots for displaying ROC curves.
# We are creating separate subplots for each probabilistic model.
# figsize=(12, 6) sets the size of the figure.
fig, axes = plt.subplots(nrows=1, ncols=n_models, figsize=(8, 6))

# Flatten the axes array if there's only one model to avoid indexing errors.
if n_models == 1:
    axes = [axes]

# Loop through each probabilistic model in the 'probabilistic_models' dictionary along with its index (i).
for i, (name, model) in enumerate(probabilistic_models.items()):
    # Get predicted probabilities for each class for the validation set using model.predict_proba(X_val_clf).
    y_prob = model.predict_proba(X_val_clf)

    # Get the number of classes in the Iris dataset.
    n_classes = len(iris_dataset.target_names)

    # Select the current subplot from the 'axes' array.
    ax = axes[i]

    # Loop through each class to create a one-vs-rest ROC curve for the current model.
    for class_index in range(n_classes):
        # Get the true binary labels for the current class.
        y_val_binary = (y_val_clf == class_index).astype(int)

        # Get the predicted probabilities for the current class.
        y_score = y_prob[:, class_index]

        # Calculate the False Positive Rate (fpr), True Positive Rate (tpr), and thresholds for the ROC curve.
        fpr, tpr, thresholds = roc_curve(y_val_binary, y_score)

        # Calculate the Area Under the ROC Curve (AUC).
        roc_auc = auc(fpr, tpr)

        # Plot the ROC curve for the current class on the current subplot.
        ax.plot(fpr, tpr, label=f'{iris_dataset.target_names[class_index]} (AUC = {roc_auc:.2f})')

    # Plot a diagonal line (random classifier line) in navy color and dashed style.
    ax.plot([0, 1], [0, 1], color='navy', linestyle='--')

    # Set the x-axis limits and label.
    ax.set_xlim([0.0, 1.0])
    ax.set_xlabel('False Positive Rate (FPR)')

    # Set the y-axis limits and label.
    ax.set_ylim([0.0, 1.05])
    ax.set_ylabel('True Positive Rate (TPR)')

    # Set the title of the subplot to include the model name.
    ax.set_title(f'{name} - ROC Curve')

    # Enable the legend to display labels for each ROC curve in the current subplot.
    ax.legend(loc="lower right")

# Adjust subplot parameters for a tight layout to prevent overlapping elements.
plt.tight_layout()

# Display the entire figure with all ROC curves in separate subplots.
plt.show()
```



**What are ROC Curves and AUC?**

*   **ROC Curve (Receiver Operating Characteristic Curve):**  A graph that visualizes the performance of a binary classifier as its discrimination threshold is varied. It plots the **True Positive Rate (TPR)** (also called Sensitivity or Recall) against the **False Positive Rate (FPR)** (also called 1 - Specificity) at different threshold settings.
    *   **TPR (True Positive Rate)**:  Out of all actual positive instances, how many were correctly classified as positive? (TP / (TP + FN))
    *   **FPR (False Positive Rate)**: Out of all actual negative instances, how many were incorrectly classified as positive? (FP / (FP + TN))
    *   By varying the classification threshold, we get different TPR and FPR values, which trace out the ROC curve.

*   **AUC (Area Under the ROC Curve):**  A single scalar value that summarizes the overall performance of a binary classifier. It represents the area under the ROC curve.
    *   **AUC ranges from 0 to 1.**
    *   **AUC = 0.5:**  A classifier with AUC = 0.5 performs no better than random guessing (its ROC curve is the diagonal line).
    *   **AUC > 0.5:**  Indicates performance better than random guessing.
    *   **AUC = 1.0:**  Represents a perfect classifier (achieves 100% TPR and 0% FPR simultaneously).
    *   **Higher AUC is generally better.**  A higher AUC means the classifier is better at distinguishing between positive and negative classes across different thresholds.

**In this code, we've adapted ROC curves for our multi-class Iris problem using a One-vs-Rest approach. For each flower type, we're essentially asking: "How well can the model distinguish this flower type from all other flower types?"  The ROC curves and AUC values help us understand the classification performance for each flower type and compare the models in terms of their ability to discriminate between classes, especially when considering different trade-offs between True Positives and False Positives.**

This is particularly useful when you care about the balance between correctly identifying positive cases and minimizing false alarms (false positives). ROC curves and AUC provide a more nuanced view of classifier performance than just accuracy, especially in situations where class imbalance or different costs for false positives and false negatives are important.

# Understanding Regularization
---
**What is Regularization?**

Regularization techniques prevent **overfitting** in machine learning models. They add constraints or penalties during training to discourage overly complex models, promoting simplicity and better generalization.

**Why is Regularization Needed?**

Regularization addresses overfitting, where models learn training data too well, including noise, leading to poor performance on new, unseen data. Regularization builds models that are:

*   **More Generalizable:** Perform well on new data.
*   **Less Sensitive to Noise:** Less affected by irrelevant data fluctuations.
*   **More Robust:** More reliable in real-world scenarios.

**For which models is Regularization used?**

Regularization is widely used in both **regression and classification**, including:

*   Linear Models (Linear & Logistic Regression)
*   Polynomial Regression
*   Decision Trees
*   Neural Networks
*   Support Vector Machines (SVMs)

**C, L1, and L2 in Logistic Regression:**

In `sklearn.linear_model.LogisticRegression`, **C**, **L1**, and **L2** relate to regularization penalties.

*   **L1 Regularization (Lasso):** (`penalty='l1'`)
    *   Penalty: Proportional to the **absolute value of coefficients**.
    *   Effect: **Drives coefficients to zero**, enabling **feature selection** and creating sparse, interpretable models.

*   **L2 Regularization (Ridge):** (`penalty='l2'`, default)
    *   Penalty: Proportional to the **square of coefficients**.
    *   Effect: **Shrinks coefficients towards zero** (but rarely to zero), reducing overfitting and improving generalization.

*   **C (Inverse Regularization Strength):**
    *   Hyperparameter controlling regularization strength (inverse relationship).
    *   **Smaller C:** Stronger regularization (larger penalty).
    *   **Larger C:** Weaker regularization (smaller penalty, approaches no regularization).

**Value Ranges and Usage of C, L1, and L2:**

*   **C:**
    *   Positive float.
    *   **Typical Range:**  e.g., `[0.001, 0.01, 0.1, 1, 10, 100, 1000]` (logarithmic scale).
    *   **Tuning C:** Optimal value found via cross-validation.

*   **L1 & L2:**
    *   Penalty *types*, not parameters with ranges. Choose one or none.
    *   Strength controlled by **C**.

**Why Use C, L1, and L2?**

*   **C (Tune Regularization):**  Optimize regularization strength to balance bias and variance for best generalization.
*   **L1 (Feature Selection):**  For models with potentially irrelevant features; creates sparse, interpretable models.
*   **L2 (Overfitting Prevention):** General-purpose overfitting reduction; improves stability and generalization, especially with multicollinearity. Often a good default choice.

**In summary, Regularization is crucial for building generalizable models. L1 and L2 penalties, tuned by C, offer distinct ways to control model complexity and prevent overfitting in linear models and beyond.**

# Loss Function with L1 and L2 Regularization
---
**Understanding the Basics**

* **Loss Function:** Measures the error between a model's predictions and the actual target values. The goal of training is to minimize this loss.
* **Regularization:** Prevents overfitting by adding a penalty to the loss function, discouraging overly complex models.
* **L1 Regularization (Lasso):** Adds a penalty proportional to the absolute value of the model's coefficients, encouraging sparsity (driving some coefficients to zero).
* **L2 Regularization (Ridge):** Adds a penalty proportional to the square of the model's coefficients, shrinking them towards zero but rarely to exactly zero. It improves generalization and stability.

**Formulas**

## Understanding Loss Functions with Regularization

Regularization techniques modify the loss function to prevent overfitting. Here's how L1 and L2 regularization affect the loss function:

**1. Loss Function with L1 Regularization:**

* **Original Loss Function (e.g., MAE):**  `MAE = (1/n) * Σ|yᵢ - ŷᵢ|`
    Where:
        * `n`: Number of data points
        * `yᵢ`: Actual target value
        * `ŷᵢ`: Predicted target value
* **L1 Penalty:** `L1 Penalty = λ * Σ|wᵢ|`
    Where:
        * `λ` (lambda): Regularization parameter controlling penalty strength. Higher values mean stronger regularization.
        * `wᵢ`: The i-th coefficient of the model.
* **Total Loss:** `Total Loss = Original Loss Function + L1 Penalty = (1/n) * Σ|yᵢ - ŷᵢ| + λ * Σ|wᵢ|`


**2. Loss Function with L2 Regularization:**

* **Original Loss Function (e.g., MSE):** `MSE = (1/n) * Σ(yᵢ - ŷᵢ)²`
    Where:
        * `n`: Number of data points
        * `yᵢ`: Actual target value
        * `ŷᵢ`: Predicted target value
* **L2 Penalty:** `L2 Penalty = λ * Σwᵢ²`
    Where:
        * `λ` (lambda): Regularization parameter controlling penalty strength. Higher values mean stronger regularization.
        * `wᵢ`: The i-th coefficient of the model.
* **Total Loss:** `Total Loss = Original Loss Function + L2 Penalty = (1/n) * Σ(yᵢ - ŷᵢ)² + λ * Σwᵢ²`


**In essence:**

* **L1 regularization** adds a penalty proportional to the absolute values of the model's coefficients.
* **L2 regularization** adds a penalty proportional to the squared values of the model's coefficients.

These penalties encourage the model to keep its coefficients small, which helps prevent overfitting by discouraging overly complex models.


**In Summary**

L1 and L2 regularization terms are added to the original loss function to create a regularized loss function. The regularization parameter (λ) controls the penalty strength, balancing fitting the training data and preventing overfitting.

In this exercise, we will explore **L2 regularization** (also known as Ridge regularization) in the context of **Logistic Regression** for Iris flower classification. L2 regularization is a common technique to prevent overfitting in linear models.

**Formulas for L2 Penalty:**

This penalty is added to the loss function during training, with $\lambda$ as the regularization parameter (in `sklearn`, $C = 1/\lambda$).

*   **L2 Penalty (Ridge):** $$ L2\ Penalty = \lambda \sum_{i=1}^{n} w_i^2 $$

**(Where $w_i$ are model coefficients)**

**Understanding L2 Regularization:**

L2 regularization adds a penalty term to the loss function during model training. This penalty is proportional to the **square of the magnitude of the coefficients** of the features.  The effect of L2 regularization is to **shrink the coefficients** towards zero. This discourages the model from giving too much weight to any single feature, effectively simplifying the model and reducing its tendency to overfit.

In `sklearn.linear_model.LogisticRegression`, L2 regularization is the **default** penalty, controlled by the parameter `C`.

*   **`C` (Inverse of Regularization Strength):**  `C` is the inverse of the regularization strength.
    *   **Smaller values of `C` imply stronger regularization.**  A smaller `C` increases the penalty for large coefficients, forcing them to shrink more towards zero.
    *   **Larger values of `C` imply weaker regularization.** As `C` approaches infinity, regularization becomes negligible, and the model behaves more like standard Logistic Regression without regularization.

**Exercise Tasks:**

1.  **Experiment with different values of `C` in `LogisticRegression`:** We will train Logistic Regression models with varying `C` values to observe the impact of L2 regularization strength on model performance.
2.  **Evaluate model performance:**  We will use Accuracy, Classification Reports, and Confusion Matrices to assess how L2 regularization affects the classification performance on the Iris validation set.

Let's implement this exercise in the code cell below.

```python
# Code Cell 16: Regularization Exercise - Classification - L2 Regularization (Logistic Regression - Ridge)

# Print a header to indicate the start of Regularization Exercise - Classification with L2 Regularization (Logistic Regression).
print("\n\n----- Step 12: Regularization Exercises - Classification - L2 Regularization (Logistic Regression - Ridge) -----")

# Define a dictionary 'classification_l2_models' to store LogisticRegression models with varying C values (L2 regularization strength).
classification_l2_models = {
    "LR (C=1e9) - No L2 Regularization (Weak)": LogisticRegression(C=1e9, penalty='l2', solver='liblinear', random_state=0, max_iter=1000), # Very large C, weak regularization
    "LR (C=1.0) - L2 Regularization (Default)": LogisticRegression(C=1.0, penalty='l2', solver='liblinear', random_state=0, max_iter=1000), # Default C=1.0, moderate regularization
    "LR (C=0.01) - L2 Regularization (Strong)": LogisticRegression(C=0.01, penalty='l2', solver='liblinear', random_state=0, max_iter=1000) # Small C=0.01, strong regularization
}

# Initialize an empty dictionary 'classification_l2_results' to store evaluation metrics for each model.
classification_l2_results = {}

# Loop through each model in the 'classification_l2_models' dictionary.
for name, model in classification_l2_models.items():
    # Train the current Logistic Regression model using the classification training data.
    model.fit(X_train_clf, y_train_clf)

    # Make predictions on the classification validation data.
    predictions_l2_clf = model.predict(X_val_clf)

    # Calculate accuracy score for the current model.
    accuracy_l2 = accuracy_score(y_val_clf, predictions_l2_clf)

    # Store accuracy, classification report, and confusion matrix in 'classification_l2_results'.
    classification_l2_results[name] = {
        "Accuracy": accuracy_l2,
        "Classification Report": classification_report(y_val_clf, predictions_l2_clf, target_names=iris_dataset.target_names),
        "Confusion Matrix": confusion_matrix(y_val_clf, predictions_l2_clf)
    }

    # Print the model name and its accuracy score.
    print(f"Model: {name}, Accuracy: {accuracy_l2:.2f}")

# Print a header to display the accuracy results in a table.
print("\n--- Classification Model Performance (Accuracy) - L2 Regularization (Logistic Regression) ---")

# Create a pandas DataFrame to display accuracy scores in a table format.
df_classification_l2_accuracy = pd.DataFrame.from_dict(classification_l2_results, orient='index', columns=['Accuracy'])

# Print the DataFrame.
print(df_classification_l2_accuracy[['Accuracy']]) # Display only Accuracy column

# Print Classification Reports for each model
print("\n--- Classification Reports - L2 Regularization (Logistic Regression) ---")
for name, results in classification_l2_results.items():
    print(f"\n--- {name} Classification Report ---")
    print(results["Classification Report"])

# Visualize Confusion Matrices for L2 Regularization (Logistic Regression)

# Create a figure and subplots for visualizing confusion matrices of models with different L2 regularization strengths.
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6)) # 1 row, 3 columns for 3 models
axes = axes.flatten() # Flatten axes array for easier indexing

# Loop through models again to create confusion matrix heatmaps.
for i, (name, results) in enumerate(classification_l2_results.items()):
    cm_l2 = results["Confusion Matrix"] # Get confusion matrix
    ax = axes[i] # Select subplot
    sns.heatmap(cm_l2, annot=True, fmt="d", cmap="Blues", ax=ax, # Heatmap
                xticklabels=iris_dataset.target_names,
                yticklabels=iris_dataset.target_names)
    ax.set_xlabel('Predicted labels') # X-axis label
    ax.set_ylabel('True labels') # Y-axis label
    ax.set_title(f'{name} - Confusion Matrix') # Subplot title

plt.tight_layout() # Adjust layout
plt.show() # Show plot

# Visualize Coefficients Before and After L2 Regularization for Target Classes (Logistic Regression)

# Create a Logistic Regression model without regularization (penalty=None)
no_reg_model = LogisticRegression(penalty=None, random_state=0, max_iter=1000)
no_reg_model.fit(X_train_clf, y_train_clf)

# Get the number of target classes
num_classes = len(iris_dataset.target_names)

# Create a figure and subplots for visualizing coefficients
fig, axes = plt.subplots(nrows=num_classes, ncols=4, figsize=(24, 6 * num_classes)) # num_classes rows, 4 columns (including no reg)

# Models to visualize (including the no regularization model)
models_to_visualize = {
    "LR (No Regularization)": no_reg_model,
    "LR (C=1e9) - Weak L2": classification_l2_models["LR (C=1e9) - No L2 Regularization (Weak)"],
    "LR (C=1.0) - Default L2": classification_l2_models["LR (C=1.0) - L2 Regularization (Default)"],
    "LR (C=0.01) - Strong L2": classification_l2_models["LR (C=0.01) - L2 Regularization (Strong)"]
}

# Define a color map for C values
color_map = {
    "LR (No Regularization)": "blue",  # Blue for no regularization
    "LR (C=1e9) - Weak L2": "green",  # Green for weak regularization
    "LR (C=1.0) - Default L2": "orange",  # Orange for default regularization
    "LR (C=0.01) - Strong L2": "red"  # Red for strong regularization
}


# Loop through models to create coefficient plots for each target class.
for i, (name, model) in enumerate(models_to_visualize.items()):
    coefficients = model.coef_  # Get coefficients for all classes

    # Loop through target classes
    for class_idx in range(num_classes):
        ax = axes[class_idx, i]  # Select subplot for current class and model

        # Create bar plot of coefficients for the current class, using color from color_map
        ax.barh(range(coefficients.shape[1]), # This generates the x-axis positions for the bars, representing the indices of the features
              coefficients[class_idx, :], # This extracts the coefficients for a specific target class (class_idx) across all features. These values determine the length of the bars.
              align="center", # Centers the bars at their respective x-axis positions.
              color=color_map[name]) # Sets the color of the bars based on the model's name

        ax.set_yticks(range(coefficients.shape[1]))
        ax.set_yticklabels(iris_dataset.feature_names)
        ax.set_xlabel('Coefficient Value')
        ax.set_title(f'{name} - Coefficients for {iris_dataset.target_names[class_idx]}')

plt.tight_layout()
plt.show()
```

In this exercise, we will investigate **L1 regularization** (also known as Lasso regularization) in **Logistic Regression** for Iris classification. L1 regularization is another powerful technique to prevent overfitting, and it has an additional property: **feature selection**.

**Formulas for L1 Penalty:**

This penalty is added to the loss function during training, with $\lambda$ as the regularization parameter (in `sklearn`, $C = 1/\lambda$).

*   **L1 Penalty (Lasso):**  $$ L1\ Penalty = \lambda \sum_{i=1}^{n} |w_i| $$

**(Where $w_i$ are model coefficients)**

**Understanding L1 Regularization:**

L1 regularization adds a penalty term to the loss function that is proportional to the **absolute value of the magnitude of the coefficients**.  The key difference from L2 regularization is that L1 regularization can **drive some coefficients to exactly zero**.  When a coefficient becomes zero, the corresponding feature is effectively excluded from the model. This means L1 regularization performs **automatic feature selection**, making the model simpler and potentially more interpretable by focusing on the most important features.

To use L1 regularization in `sklearn.linear_model.LogisticRegression`, we need to set the `penalty` parameter to `'l1'` and choose a suitable solver that supports L1 penalty (like `'liblinear'`, `'saga'`, or `'liblinear'`).  Again, the `C` parameter controls the regularization strength (inverse relationship).

```python
# Code Cell 18: Impact of Various C Values on Logistic Regression ---

# 1. Define a range of C values
C_values = [1e9, 10, 1, 0.1, 0.01]  # Varying regularization strengths

# 2. Create a dictionary to store models and results
classification_c_models = {}
classification_c_results = {}

# 3. Train and evaluate models for each C value
for C in C_values:
    model_name = f"LR (C={C})"
    model = LogisticRegression(C=C, penalty='l1', solver='liblinear', random_state=0, max_iter=1000)
    model.fit(X_train_clf, y_train_clf)
    predictions = model.predict(X_val_clf)
    accuracy = accuracy_score(y_val_clf, predictions)

    classification_c_models[model_name] = model
    classification_c_results[model_name] = {
        'Accuracy': accuracy,
        'Classification Report': classification_report(y_val_clf, predictions),
        'Confusion Matrix': confusion_matrix(y_val_clf, predictions)
    }

    print(f"Model: {model_name}, Accuracy: {accuracy:.2f}")

# --- Visualize Confusion Matrices ---

# 1. Create subplots for confusion matrices
num_models = len(C_values)
fig, axes = plt.subplots(nrows=1, ncols=num_models, figsize=(4 * num_models, 4))  # Adjust figure size as needed

# 2. Plot confusion matrices for each model
for i, (name, results) in enumerate(classification_c_results.items()):
    cm = results['Confusion Matrix']
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i], cbar=False)  # Customize as needed
    axes[i].set_xlabel('Predicted Label')
    axes[i].set_ylabel('True Label')
    axes[i].set_title(f'Confusion Matrix - {name}')

plt.tight_layout()
plt.show()

# --- Visualize the Impact of C on Coefficients (Bar Plots for All Classes) ---

num_classes = len(iris_dataset.target_names)  # Get the number of target classes

for class_idx in range(num_classes):
    coefficients_c = {}
    for name, model in classification_c_models.items():
        coefficients_c[name] = model.coef_[class_idx]  # Get coefficients for the current class

    df_coefficients_c = pd.DataFrame(coefficients_c, index=iris_dataset.feature_names)

    # Bar plot for the current class
    ax = df_coefficients_c.plot(kind='bar', figsize=(12, 6))
    ax.set_xlabel('Features')
    ax.set_ylabel('Coefficient Value')
    ax.set_title(f'Impact of C on Logistic Regression Coefficients (Class: {iris_dataset.target_names[class_idx]})')
    ax.legend(title='Regularization Strength (C)')
    plt.xticks(rotation=0)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

## How to Interpret the Bar Plot:

- **Y-axis:** Represents the coefficient values for each feature.

Random Forests are a powerful ensemble method that builds upon the Decision Tree concept and provides an elegant and effective way to combat overfitting and improve model robustness even further.

Recall that in this module, we saw how individual Decision Trees, especially when grown without constraints, can easily overfit. **Random Forests address this overfitting tendency by creating *ensembles* of many Decision Trees.**  These trees are built in a randomized way (using techniques like bootstrap aggregating - "bagging" - and random feature selection), and their predictions are then combined (averaged for regression, majority vote for classification).

**Key advantages of Random Forests that directly relate to our discussion of underfitting and overfitting:**

*   **Reduced Overfitting:** By averaging predictions from many trees, Random Forests significantly reduce the variance and overfitting tendencies of individual Decision Trees.  The "wisdom of the crowd" effect helps to cancel out the noise learned by individual trees.
*   **Improved Generalization:** Random Forests typically generalize much better to unseen data compared to single Decision Trees, leading to more robust and reliable models in practice.
*   **High Accuracy and Robustness:** Random Forests often achieve high accuracy and are less sensitive to hyperparameter tuning compared to single Decision Trees.
